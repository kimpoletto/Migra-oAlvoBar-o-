```python
# diretorio que estou procurando os arquivos do ALvo BARA
diretorio = 'Digital_CEDOC\\MLUA\\Fotos_Figuras\\Sondagem'

diretorio_saida = '@migracao\\04_CORE_SHED\\SOUTH_HUB\\MLUA\\Drill_Logs'

tema = 'Core_Photos'

# Lista de arquivos

for root, dirs, files in os.walk(diretorio):
    contador = 1
    for arquivo in files:
        try: 

            if '\\act' in root:
                continue
            if '.7z' in arquivo:
                pass
            else:

                a = arquivo.replace('pkc_','')
                b = a.replace('pso_','')
                c = b.replace('prof','prof_')
                d = c.replace('cx','bx')
                e = d.replace('molhada','wet')
                f = e.replace('seca','dry')
                g = f.replace('m','_m')
                h = g.replace('__','_')
                i = h.replace('_mlua','mlua')
                k = i.replace('m_wet','wet')
                l = k.replace('m_dry','dry')
                m = l.replace('wet','m_wet')
                n = m.replace('dry','m_dry')

                nome_arquivo = f'{n}'

                j = n.split('_')[1]



                pasta_furo = f'PKC-MLUA-{j.upper()}'

                print(f'{pasta_furo}')
                print(f'{nome_arquivo}')
                
                o = f'{diretorio_saida}\\{pasta_furo}\\{tema}\\{nome_arquivo}'
                saida = o.replace('/','\\')
                
                
                shutil.move(f'{root}\\{arquivo}', f'{saida}')
                contador += 1

        except:
            print(f'ERRO: {root}/{arquivo}')
        
print(contador)

```
